```python
import pandas as pd

url ='http://www.bjets.com.cn/article/jyxx/?'
links = []
for n in range(1,62):
    link = url + str(n)
    links.append(link)
    
df2 = pd.DataFrame()
for url in links:
    dfs = pd.read_html(url,header=0)
    for df in dfs:
        df2 = df2.append(df,ignore_index=True)
        
df2.to_excel('test1.xlsx',encoding='utf_8')
```
